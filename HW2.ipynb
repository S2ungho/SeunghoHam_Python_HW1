```python
#단순연결리스트 응용예제 02 로또 추첨하기
import random

class Node():
    def __init__(self):
        self.data = None
        self.link = None
        
def printNodes(start):
    current = start
    if current == None:
        return
    print(current.data, end=' ')
    while current.link != None:
        current = current.link
        print(current.data, end = ' ')
    print()
    
head, current, pre = None, None, None
lotto = []
##메인 코드 

if __name__ == "__main__":
    
    lotto = random.sample(range(1,46),6)
    node = Node()
    node.data = lotto[0]
    head = node
        
    for data in lotto[1:]:
        pre = node
        node = Node()
        node.data = data
        pre.link = node

    printNodes(head)
```

```python
#스택 수업자료 응용예제 02 파일 내용을 완전히 거꾸로 출력하기

print("----------------before-----------------")
print("""진달래꽃 
나 보기가 역겨워 
가실 때에는 
말없이 고이 보내드리오리다.""")

print("-------------after-------------")
before ="""진달래꽃 
      나 보기가 역겨워 
      가실 때에는 
      말없이 고이 보내드리오리다."""
      
SIZE = len(before)
stack = [None for _ in range(SIZE)]
top = -1

for data in before:
    top += 1
    stack[top] = data
    
for data in stack:
    print(stack[top],end = '')
    stack[top] = None
    top -= 1
    
```

```python
#큐 수업자료 응용예제 02 콜센터의 응답 대기 시간 계산하기
def isQueueFull():
    global SIZE, queue,front,rear
    if ((rear+1)%SIZE==front):
        return True
    else:
        return False
    
def isQueueEmpty():
    global SIZE, queue,front,rear
    if (front==rear):
        return True
    else:
        return False
    
def enQueue(data):
    global SIZE, queue,front,rear
    if (isQueueFull()==True):
        print("큐가 꽉 찼습니다.")
        return
    rear=(rear+1)%SIZE
    queue[rear]=data
    
def deQueue():
    global SIZE, queue,front,rear
    if (isQueueEmpty()==True):
        print("큐가 비었습니다.")
        return None
    front=(front+1)%SIZE
    data=queue[front]
    queue[front]=None
    return data


SIZE = 6
queue = [None for _ in range(SIZE)]
front = rear = 0
sec = 0
cnt = 0
call = [('사용', 9),('고장', 3), ('환불', 4), ('환불', 4), ('고장', 3)]
#메인 코드

if __name__ == "__main__":
    
    for data in call:
        if queue[cnt] != None:
            sec += queue[cnt][1]
        print("귀하의 대기 예상시간은 ",sec ,"분 입니다.")
        print("현재 대기 콜 --> ", queue)
        print()
        enQueue(data)
        cnt+=1
        
    print("최종 대기 콜 -> ",queue)
    print("프로그램 종료!")
    
```
